## Program untuk 1 folder

```python
import os
import cv2
import shutil
import numpy as np
```

```python
count = 0
folder_image = r"D:\SCALPING\tradebot_mirae\klasifikasi arah sinyal\photos"
folder_naik = r"D:\SCALPING\tradebot_mirae\klasifikasi arah sinyal\naik"
images = os.listdir(folder_image)
for image in images:
    img_path = f"{folder_image}/{image}"
    img = cv2.imread(img_path)
    y = 685
    x = 550
    h = 55
    w = 20
    crop_img = img[y:y+h, x:x+w]
    gray = cv2.cvtColor(crop_img, cv2.COLOR_BGR2GRAY)
    ret, bin_img = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY)
    tinggi_1 = 0
    tinggi_2 = 0

    for enum, i in enumerate(range(bin_img.shape[0]-1, 0, -1)):
        if bin_img[i,3] < 10:
            tinggi_1 = enum
            break
        
    for enum, i in enumerate(range(bin_img.shape[0]-1, 0, -1)):
        if bin_img[i,15] < 10:
            tinggi_2 = enum
            break

    if tinggi_2 > tinggi_1: 
        count+=1
        img_naik = f"{folder_naik}/{count}.jpg"
        shutil.copyfile(img_path, img_naik)
print(count)
```
